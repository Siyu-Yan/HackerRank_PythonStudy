# Polar Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1 
theta = np.linspace(0, 2*np.pi, 1000)
r = a *np.sin(12*theta)
plt.axes(polar = True)
plt.plot(theta,r, '#81d8d0',':')
plt.show()
```

```python
a = 1
theta = np.linspace(0, 2*np.pi, 1000)
r = np.sqrt(2*(a**2)*np.cos(10*theta))
plt.axes(polar = True)
plt.plot(theta, r)
plt.show()
```

## 把复数变成极坐标

```python
cmath.polar(1 + 1j) 
 #输出: (1.4142135623730951, 0.7853981633974483) 
# == (sqrt(1 + 1), atan2(1, 1))
# 在极坐标中返回 x 的表达方式。返回一个数对 (r, phi)，r 是 x 的模数，phi 是 x 的相位角 
# polar(x) 相当于 (abs(x), phase(x))。
```

```python
cmath.polar(complex(-1.0, 0.0)) 
# 在极坐标中返回 x 的表达方式。返回一个数对 (r, phi)，r 是 x 的模数，phi 是 x 的相位角 
# polar(x) 相当于 (abs(x), phase(x))。
```

# Triangle

```python
for i in range(1,6):
    l = ""
    for j in range(1, i + 1):
        l += str(j)
    for j in range(i - 1, 0, -1):
        l += str(j)
    print(l)
```

```python
for i in range(1,5+1):
    print (((10**(i) - 1) // 9 )**2)
```
