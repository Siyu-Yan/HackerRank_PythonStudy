# Transpose and Flatten

## Transpose

We can generate the transposition of an array using the tool numpy.transpose.

It will not affect the original array, but it will create a new array.

```python
my_array = numpy.array([[1,2,3],
                        [4,5,6]])
print(numpy.transpose(my_array))
```

```python
print(my_array.transpose())
```

## Flatten
The tool flatten creates a copy of the input array flattened to one dimension.

```python
my_array = numpy.array([[1,2,3],
                        [4,5,6]])
print(my_array.flatten())
```

```python
print(my_array.reshape(-1))
```

# Eye and Identity

identity

The identity tool returns an identity array. An identity array is a square matrix with all the main diagonal elements as 1 and the rest as 0. The default type of elements is float.

The eye tool returns a 2-D array with 1's as the diagonal and 0's elsewhere. The diagonal can be main, upper or lower depending on the optional parameter k. A positive k is for the upper diagonal, a negative k is for the lower, and a 0 k (default) is for the main diagonal.

```python
numpy.identity(3) #3 is for  dimension 3 X 3
```

```python
numpy.eye(8, 7, k = 1)    # 8 X 7 Dimensional array with first upper diagonal 1.
```

# Sum and Prod

The sum tool returns the sum of array elements over a given axis.

The prod tool returns the product of array elements over a given axis.

```python
my_array = numpy.array([ [1, 2], [3, 4] ])

print(numpy.sum(my_array, axis = 0)) # 0 is by column        
print(numpy.sum(my_array, axis = 1)) #by row    
print(numpy.sum(my_array, axis = None)) # all numbers
print(numpy.sum(my_array))   
```

```python
import numpy

my_array = numpy.array([ [1, 2], [3, 4] ])

print(numpy.prod(my_array, axis = 0))          
print(numpy.prod(my_array, axis = 1)) 
print(numpy.prod(my_array, axis = None))  
print(numpy.prod(my_array))
```

# Polynomials

The poly tool returns the coefficients of a polynomial with the given sequence of roots.

```python
numpy.poly([-1, 1, 1, 10]) 
# Return: 1D array having coefficients of the polynomial from the highest degree to the lowest one.
# c[0] * x**(N) + c[1] * x**(N-1) + … + c[N-1] * x + c[N] where c[0] always equals 1.
```

```python
numpy.roots([1, 0, -1]) # 方程求解
```

```python
numpy.polyint([1, 1, -1]) # 积分
```

```python
numpy.polyder([1, 1, 1, 1])  # 一阶微分
```

```python
numpy.polyval([1, -2, 0, 2], 4) # 赋值4 4^3-2*4^2 +2
```

```python
numpy.polyfit([0,1,-1, 2, -2], [0,1,1, 4, 4], 2)
```

# Linear Algebra

linalg.det

The linalg.det tool computes the determinant of an array.

```python
numpy.linalg.det([[1 , 2], [2, 1]])  # return The deteminant of that square array. 行列式 2*2-1
```

linalg.eig

The linalg.eig computes the eigenvalues and right eigenvectors of a square array.

```python
vals, vecs = numpy.linalg.eig([[1 , 2], [2, 1]])
print(vals)                                 
print(vecs)     # 特征值
```

linalg.inv

The linalg.inv tool computes the (multiplicative) inverse of a matrix.

```python
numpy.linalg.inv([[1 , 2], [2, 1]])    # 逆矩阵
```
