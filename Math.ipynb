# Polar Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1 
theta = np.linspace(0, 2*np.pi, 1000)
r = a *np.sin(12*theta)
plt.axes(polar = True)
plt.plot(theta,r, '#81d8d0',':')
plt.show()
```

```python
a = 1
theta = np.linspace(0, 2*np.pi, 1000)
r = np.sqrt(2*(a**2)*np.cos(10*theta))
plt.axes(polar = True)
plt.plot(theta, r)
plt.show()
```

# Triangle

```python
for i in range(1,6):
    l = ""
    for j in range(1, i + 1):
        l += str(j)
    for j in range(i - 1, 0, -1):
        l += str(j)
    print(l)
```
